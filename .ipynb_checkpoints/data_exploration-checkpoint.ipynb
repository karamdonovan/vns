```python
# read in the data
trials = pd.read_hdf('C:/Users/karam/Documents/GitHub/1-normalized.hdf','trials')
eye_samples = pd.read_hdf('C:/Users/karam/Documents/GitHub/1-normalized.hdf','eye_samples')
```
